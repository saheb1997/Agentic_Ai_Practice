```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embedding=(
    OllamaEmbeddings(model="codellama")
    
)
```

```python
r1=embedding.embed_documents(
    [
        "Hello, how are you?",
        "I am fine, thank you.",
        "What is your name?",
        "My name is Ollama.",
        "I am an AI model.",
        "I can help you with your tasks.",
        "What do you want to know?",
        "I can answer your questions.",
        "I can also generate text.",
        "I can help you with your projects."
    ]
)
r1
```

```python
r2=embedding.embed_query("how you father are?")
```
