```python
from langchain_openai import OpenAIEmbeddings
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
embeddings
```

```python
text = "This is tutotial on oepnai embeding"
query_result = embeddings_1024.embed_query(text)
query_result
```

```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader("/Users/sahebsarkar/Desktop/projects/Agentic_Ai_Practice/Data_ingesion/spech.txt")
docs=loader.load()
docs
```

## Text_spliter is used


```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_spliter = RecursiveCharacterTextSplitter(chunk_size = 500 , chunk_overlap=500)
final_document = text_spliter.split_documents(docs)
final_document
```

```python
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(final_document,embeddings_1024)
db
```

```python
###retrive the result from query vectorstore db
query="Generative models, such as Generative Adversarial Networks (GANs) and large language models, can create new content, including images, music, and text."
retrive_result = db.similarity_search(query)
print(retrive_result)
```
